Evaluate the gradient of the following function w.r.t. the input:

$$ f(x,y,z) =  log(\sigma(\frac{cos(\pi \times x)+sin(\pi \times y/2)}{tanh(z^2)}))$$
where $\sigma$ is the sigmoid function. Find gradient for the following inputs:
  * $(x,y,z)$ =  (1,2,3)
  * $(x,y,z)$ =  (3,2,1)
  * $(x,y,z)$ =  (12,23,31)
  * $(x,y,z)$ =  (32,21,13)

Using the above classes, construct the function:

$$ f(x,y,z) =  log(\sigma(\frac{cos(\pi \times x)+sin(\pi \times y/2)}{tanh(z^2)}))$$

and write the forward computation to calculate $$ f(x, y, z) $$ and also comute the partial derivative w.r.t all the input variables $$ \frac{\partial{f}}{\partial{x}}, \frac{\partial{f}}{\partial{y}}, \frac{\partial{f}}{\partial{z}} $$

```python
# Now write the class func
# which constructs the graph (all operators), forward and backward functions.

class Func():
  def __init__(self):
    # construct the graph here
    # assign the instances of function modules to self.var
    self.node_cos = Cosine()
    self.node_sin = Sine()
    self.node_tanh = Tanh()
    self.sigmoid = Sigmoid()
    self.log = Log()
    self.var=None
    self.x=None
    self.y=None
    self.z=None
    self.var_result=None
    self.common_termx=None
    self.common_termy=None
    self.common_termz=None
    self.common_terme_=None

  def forward(self,x,y,z):
    # Using the graph element's forward functions, get the output.
    self.x=np.float64(x)
    self.y=np.float64(y)
    self.z=np.float64(z)
    self.common_termx=self.node_cos.forward(np.pi*self.x)
    self.common_termy=self.node_sin.forward(np.pi*self.y/2)
    self.common_termz=self.node_tanh.forward(self.z**2)
    self.common_terme=(self.common_termx+self.common_termy)/self.common_termz
    self.common_terme_=-(self.common_termx+self.common_termy)/self.common_termz
    self.var_result=self.log.forward(self.sigmoid.forward(self.common_termx+self.common_termy)/self.common_termz)
    print("Final Function value at x={},y={},z={} is : {}".format(x,y,z,self.var_result))
    return None

  def backward(self):
    # Use the saved outputs of each module, and backward() function calls

    combined_term=(1+self.common_terme_)*-self.common_terme_
    grad_x=self.node_cos.backward(combined_term*np.pi/self.common_termz)
    grad_y=self.node_sin.backward(combined_term*-np.pi/(2*self.common_termz))
    grad_z=self.node_tanh.backward(combined_term*2*self.z*(self.common_termx+self.common_termy)/self.common_termz**2)
    return [grad_x,grad_y,grad_z]
```

```python
inputs=[(1.0,2.0,3.0),(3.0,2.0,1.0),(12.0,23.0,31.0),(32.0,21.0,13.0)]
print("\n\n\n ################################################################ Function Output ################################################################")
for (x,y,z) in inputs:
  print("Input Values x={},y={},z={}".format(x,y,z))
  calc_grad=Func()
  calc_grad.forward(x,y,z)

  #print("\n################################################################ Gradient Output ################################################################")
  a=calc_grad.backward()
  print("Gradient with respect to x = ",a[0])
  print("Gradient with respect to y = ",a[1])
  print("Gradient with respect to z = ",a[2])
  print("\n\n\n\n\n")
```
